## Jupyter notebook
The jupyter notebook is an interactive computing environment to:

- create and share documents mixing code, text, visualisations
- step-by-step interact with **data** and **code**
- document data analysis
- **test** code ideas

 It is not used for:

 - used for **developing** code
 - code documentation

**Advantages**
- Excellent for Data Science Workflows
- Support mixing markdown, python, R, raw code, ...
- Great for showcasing and tutorials
- Easy sharing and collaboration

**Disadvantages**
- Not great to construct **reproducable (automatic) workflows**
- Not suited for production code
- Difficult version control (opposed to code in git)
- Encourages poor practices when used incorrectly
    - Duplicate code
    - Polluted state variables
    - **Weak** structure
    - Hard to debug

## Operations

We can also add, substract and alter values. Maybe not sensible, but let's compute the percentage dry yield compared to Fresh yield

```python
(df["Dry_Yield"]/df["Fresh_Yield"])
```

Oh no, I'm getting a lot of NaN's, and I don't want NaN's. Have a close look: apparently for many records dry and wet yield are not present. We can select the records that are not equal to NaN

```python
quotient = (df["Dry_Yield"]/df["Fresh_Yield"])
```

```python
quotient[~quotient.isnull()]
```

That seems better. Let's have a look at the statistics 

```python
quotient[~quotient.isnull()].describe()
```

You can also substract

```python
substract = df["Fresh_Yield"]-df["Dry_Yield"]
```

```python
substract[~substract.isnull()]
```
